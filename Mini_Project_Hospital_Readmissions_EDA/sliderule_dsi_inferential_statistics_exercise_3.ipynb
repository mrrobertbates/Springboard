# Hospital Readmissions Data Analysis and Recommendations for Reduction

### Background
In October 2012, the US government's Center for Medicare and Medicaid Services (CMS) began reducing Medicare payments for Inpatient Prospective Payment System hospitals with excess readmissions. Excess readmissions are measured by a ratio, by dividing a hospital’s number of “predicted” 30-day readmissions for heart attack, heart failure, and pneumonia by the number that would be “expected,” based on an average hospital with similar patients. A ratio greater than 1 indicates excess readmissions.

I think the claim that hospital capacity and readmission rates are correlated may be unfounded.  The graph does suggest that the two may be correlated, but only slightly so.  The statistics cited to support this claim are that hospitals with less than 100 discharges have a mean excess readmission rate of 1.023 with 63% of such hospitals having a readmission rate greater than 1, while hospitals with more than 1000 discharges have a mean excess readmission rate of 0.978 with 44% of such hospitals have a readmission rate of less than 1.

We will need to determine if these statistics are significant by means of 2-sample significance tests.  Lets chose $\alpha$ = 0.01 and take the null hypothese to be that the excess readmission rates for hospitals with less than 100 discharges and the excess readmission rates for hispitals with more than 1000 discharges are not different.

We will evaluate the null hypothesis by means of the sample mean test statistic.

```python
clean_hospital_read_df.head()
```

```python
df = clean_hospital_read_df.loc[clean_hospital_read_df["Excess Readmission Ratio"].notnull()]
```

```python
df_under_100 = df.loc[df["Number of Discharges"] < 100]
```

```python
df_over_1000 = df.loc[df["Number of Discharges"] > 1000]
```

```python
df_under_100["Excess Readmission Ratio"].count()
```

```python
df_over_1000["Excess Readmission Ratio"].count()
```

```python
diff_mean = df_under_100["Excess Readmission Ratio"].mean() - df_over_1000["Excess Readmission Ratio"].mean()
diff_mean
```

Lets assume the null hypothesis is true and simulate the condition that the null hypothesis is true by pooling the data.

```python
pooled_data = pd.concat([df_under_100["Excess Readmission Ratio"], df_over_1000["Excess Readmission Ratio"]], axis=0)
```

```python
under_100_replicate = np.empty(10000)
over_1000_replicate = np.empty(10000)
for each in range(0, 10000):
    pooled_permutation = np.random.permutation(pooled_data)
    under_100_replicate[each] = np.mean(pooled_permutation[0:1188])
    over_1000_replicate[each] = np.mean(pooled_permutation[1188:])
```

```python
p_val = np.sum((np.abs(under_100_replicate - over_1000_replicate)) >= diff_mean)/len(pooled_data)
print("The p-value is {}".format(p_val))
```

Based on the result above, if it were true that there is no difference in excess readmission rates for hospitals with less than 100 discharges and hospitals with more than 1000 discharges then it would be nearly impossible to obtain the difference in mean above.  Thus we must reject the null hypothesis in favor of the alternative, that there is a difference in excess readmission rates for hospitals with less than 100 discharges and hospitals with more than 1000 discharges.

Lets confirm these result by using more traditional statistical methods.

```python
import scipy.stats as stats
```

```python
df_under_100["Excess Readmission Ratio"].std()
```

```python
df_over_1000["Excess Readmission Ratio"].std()
```

Since the excess readmission ration for hospitals with under 100 discharges and hospitals with over 1000 discharges do not appear to have equal variances, we will to use the non-pooled variance as an estimator.

For large sample sizes we may use the test statistic $T_d = \frac{\bar{X}_n - \bar{Y}_m}{S_d}$, where $S_d$ is the non-pooled variance, $\bar{X}_n$ is the sample mean for the hospitals with less than 100 discharges, and $\bar{Y}_m$ is the sample mean for the hospitals with over 1000 discharges.  Since the sample size is large we may approximate the test statistic $T_d$ using the standard normal distribution.

```python
S_d = np.sqrt(df_under_100["Excess Readmission Ratio"].var()/1188 + df_over_1000["Excess Readmission Ratio"].var()/463)
S_d
```

```python
p_val = 2*(1-stats.norm(0,1).cdf(diff_mean/S_d))
print("The p-value is {}.".format(p_val))
```

Once again, if we assume $H_o$, that there is no difference in the excess readmission rates between hospitals with less than 100 discharges and hospitals with more than 1000 discharges, then it is nearly impossible that we should obtain a difference in sample means at least as extreme as 0.044.  This suggests that the alternative hypothesis, $H_a$, that there is a difference in the excess readmission rates is true.
