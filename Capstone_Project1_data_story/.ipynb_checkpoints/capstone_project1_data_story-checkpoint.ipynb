```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
columns = pd.read_csv('bank_altered.csv').columns[1:]
```

```python
df = pd.read_csv('bank_altered.csv', usecols=columns)
df.head()
```

```python
sns.set_style('whitegrid')
plt.boxplot(df.loc[df.y == 1].age)
plt.title("Age of Depositors")
plt.ylabel("Age")
plt.show()
```

```python
sns.set_style('whitegrid')
plt.boxplot(df.loc[df.y == 0].age)
plt.title("Age of Non-Depositors")
plt.ylabel("Age")
plt.show()
```

```python
sns.kdeplot(df.balance.values)
plt.title("Average Balance Density Estimate")
plt.xlabel("Average Balance")
plt.ylabel("Density")
```

```python
df.balance.describe()
```

```python
plt.scatter(df.loc[df.y == 0].age, df.loc[df.y == 0].balance, color = 'blue')
plt.scatter(df.loc[df.y == 1].age, df.loc[df.y == 1].balance, color = 'red')
plt.show()
```

```python
plt.boxplot(df.loc[df.y==1].duration)
plt.show()
```

```python
plt.boxplot(df.loc[df.y==0].duration)
plt.show()
```

```python
sns.kdeplot(df.loc[df.y==1].duration/60)
plt.xticks(np.arange(0, 70, 5))
plt.title("Density Estimate of Call Duration for Depositors")
plt.xlabel("Call Duration in Minutes")
plt.ylabel("Density")
plt.show()
```

```python
sns.kdeplot(df.loc[df.y==0].duration/60)
plt.xticks(np.arange(0, 90, 5))
plt.title("Density Estimate of Call Duration for Non-Depositors")
plt.xlabel("Call Duration in Minutes")
plt.ylabel("Density")
plt.show()
```
